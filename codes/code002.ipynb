# Instalar Cookiecutter



## Qu es Cookiecutter?
Es un manejador de plantillas multiplataforma (Windows, Mac OS, Linux) que te permite hacer plantillas en lenguaje de programacin (Python,Julia, R) o lenguaje de marcado (Markdown). Puede ser usado como herramienta de lnea de comandos o como librera de Python.

## [Cmo funciona?](https://cookiecutter.readthedocs.io/en/1.7.2/README.html)



[Cookiecutter](https://cookiecutter-data-science.drivendata.org/) funciona con Jinja, un motor de plantillas extensible con el cual puedes crear plantillas como si estuvieras escribiendo cdigo en Python.



Hay 3 pasos para entender la manera en que funciona:

1. Detectar una sintaxis especial en los documentos y carpetas de tu proyecto.
2. Buscar variables y valores a reemplazar.
3. Finalmente, entregar un proyecto con la estructura definida en la plantilla.

## Sintaxis de Jinja
Existen 3 tipos diferentes de bloques:

1. Bloques de expresin: se usan para incluir variables en la plantilla:           
```{{ cookiecutter.saluda }}```
2. Bloques de declaracin: se usan para el uso de condicionales, ciclos, etc.:        
``{% if coockiecutter.eres_asombroso %}``        
`...`          
`{% endif %}`
3. Bloques de comentario: se usan para dejar comentarios o recomendaciones a los propios desarrolladores:     
```{# Esto es un comentario (Recomendaciones para quien lea la plantilla) #}```



## Instalacin de Cookiecutter
**Importante:** Todos los pasos son va consola y con Anaconda instalado.

### Crea una carpeta un entrar en ella:
```sh
#Estando en /Configuracion_profesional_Entorno/codes
#Creamos una carpeta
mkdir cookiecutter-personal
cd cookiecutter-personal
```
## Instalando Cookiecutter:
### Agrega el canal Conda-Forge a tu configuracin global:
```sh
#Cookiecutter se encuentra solo en este canal
#por eso primero agregamos el canal
conda config --add channels conda-forge
```
### Crea un ambiente virtual que contenga a Coockiecutter:
```sh
#Podemos llamarlo env, para que sigamos la convencin
#de agregar a un repositorio con gitignore 
conda create --name env cookiecutter=1.7.3
#o cookiecutter
conda create --name cookiecutter cookiecutter=1.7.3
```
Con este comando conda empezara a descargar e instalar paquetes a mi entorno virtual para utilizar cookie cutter.
### Activa el ambiente virtual:
```sh
#escogimos el nombre de env
conda activate env
```
### Definir en dnde estar tu ambiente:
```sh
#Para exportar las configuraciones de un ambiente
conda env export --from-history --file environment.yml
```
### Para desactivar el ambiente virtual:
```sh
conda deactivate
```
