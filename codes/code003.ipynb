# Crear plantilla de proyecto personalizadas

## Estructura inicial de la plantilla

Para eso necesitamos ir a la terminal y vamos a posicionarnos en [cookiecutter](cookiecutter-personal/) mediante los siguientes comandos.
 ```bash
 cd cookiecutter-personal
 ```

Una vez ahí, necesitamos borrar la carpeta **cookiecutter_testing** que habíamos creado en la anterior clase con el comando del repositorio de git hub.

### Creando la estructura
Para ello abrimos VS Code con el comando `code .`, se nos abrirá. Por lo que haremos lo siguiente:

1.- La vamos a nombrar como (contendrá todo el contenido que queramos para el proyecto):
```bash
#Esto es una sintaxis de JINJA
#Ademas es un bloque de expresión
#Vamos a sustituir el valor de una variable
#Y sustituirlo
{{cookiecutter.project_slug}}
```



Dentro de la carpeta podemos definir archivos, es decir crear un archivo [README.md](cookiecutter-personal/{{cookiecutter.project_slug}}/README.md). También podemos crear carpetas.



Podemos editar el **README.md** y usar la sintaxis de JINJA para personalizar nuestro proyecto.

**Contenido de README.md**
```md
#Este es el titulo n1 del proyecto con la sintaxis JINJA
# {{ cookiecutter.project_title}}

#Aquí tenemos nombre de autor
By {{cookiecutter.author_name}}

#Este otro es un titulo n3 que tiene la descripción del proyecto
### {{cookiecutter.project_description}}
```
No solo podemos crear un **README.md** con la sintaxis de JINJA, sino que podemos crear los archivos que tengan cierta sintaxis. Por ejemplo, podemos crear un archivo [environment.yml](cookiecutter-personal/{{cookiecutter.project_slug}}/environment.yml)


En el contenido escribimos lo siguiente (hay que fijarnos en la sintaxis de JINJA):


```yaml
#Para poder crear un entorno en conda corremos
#conda env create --file environment.yml

#NOMBRE DEL PROYECTO
name: {{cookiecutter.project_slug}}
channels:
- anaconda
- conda-forge
- default
dependencies:
#BLOQUE DECLARATIVO
{% if cookiecutter.project_packages == 'All' -%}
- fs
- jupyter
- jupyterlab
- pathlib

{% endif -%}
- pip

{% if cookiecutter.project_packages == "All" -%}
- pyprojroot

{% endif -%}
- python={{ cookiecutter.python_version }}
- pip:
    {% if cookiecutter.project_packages == "All" -%}
    - pyhere
    {% endif -%}
```

Recordemos que debemos tener el orden de esta forma:

       
Dentro de la carpeta tendremos todo el arbol de carpetas, archivos **README.md**, **.gitignore** y todo lo relacionado al entorno.




## Archivo de configuración

Para poder desplegar el entorno y la plantilla creada necesitamos un archivo de configuración [cookiecutter.json](cookiecutter-personal/cookiecutter.json).

Dentro de este archivo crearemos todas las variables que utilizamos anteriormente en nuestro proyecto.

NO COPIES LAS LINEAS EN ROJO (#) son comentarios para guiarnos

```json
{   
    #(OJO si no colocas nada, existe un valor por defecto que es el que esta entre comillas)
    #titulo del proyecto 
    "project_title": "Project Title",
    #nombre del proyecto
    "project_slug": "project_personal",
    #Descripción del proyecto
    "project_description": "This is a project",
    #Nombre del autor
    "author_name":"Type your name",
    #Paquetes necesarios para el proyecto
    "project_packages":["All","Minimal"],
    #Versión de Python
    "python_version":"3.7"
}
```

Así que ya tenemos la estructura donde tenemos los siguientes archivos y carpetas:



En ella vemos que tenemos el **archivo de configuración** [](cookiecutter-personal/cookiecutter.json), que nos da los valores de las variables que establecimos por defecto
