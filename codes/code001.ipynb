# ¿Qué son y por qué utilizar plantillas de proyectos?

Vas a aprender a crear todo un sistema de carpetas y archivos que siempre que llamemos al comando se repliquen toda esta plantilla.

Ademas desarrollaremos proyectos autocontenidos y multiplataforma, lo podemos pasar en cualquier entorno.

Las plantillas son un medio que posibilita portar o construir un diseño predefinido, por ejemplo en ciencia de datos vamos a definir un conjunto de carpetas de datos, notebooks, es decir todo el contenido del proyecto. Ademas lo podemos automatizar a partir de una plantilla de proyecto.


Ya no vamos a tener que pensar en que nombre ponerle a los archivos o como estará construida la estructura de mi proyecto. Son pequeñas decisiones sencillas pero son bastante y necesitamos agilizar el proceso.

Ademas podemos personalizar nuestra estructura añadiendo mas carpetas para organizar nuestro proyecto a nuestro modo.

Como Científico de datos, tenemos que tomar en cuenta la reproducibilidad.

Una estructura y orden nos ayudará a hacer las cosas de una forma mas entendible y sencilla.

En la siguiente clase vamos a utilizar un programa que administra las plantillas. 

## Extra

Las plantillas de proyectos son un medio que posibilita portar o construir un diseño predefinido. Estas te permiten definir carpetas, notebooks, scripts, archivos de configuración, etc.

## ¿Por qué usar plantillas de proyectos?
Algunas razones para usar plantillas en proyectos se debe a que:

Agiliza tu trabajo y reduce la fatiga por decisión.
Es más fácil de personalizar un proyecto hecho con plantillas que hecho desde cero.
La reproducibilidad de un proyecto es más viable.
Es más fácil encontrar algo entre los componentes del proyecto.
