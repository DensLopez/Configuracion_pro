El manejo de rutas del sistema es esencial para trabajar con archivos y directorios en un programa. El módulo osen Python proporciona funciones para interactuar con el sistema operativo, incluida la manipulación de rutas de manera efectiva y portátil.

### Obtener la Ruta Actual:
Puedes usar `os.getcwd()` para obtener la ruta absoluta del directorio actual.

```py
import os 

current_dir = os.getcwd()
print("Directorio actual:", current_dir)
```
### Unir Rutas:
`os.path.join()` te permite unir partes de rutas en una ruta única y válida compatible con el sistema.

```py
path = os.path.join("carpeta", "subcarpeta", "archivo.txt")
print("Ruta unida:", path)
```

### Obtener Nombre de Archivo y Directorio Base:

`os.path.basename()` devuelve el nombre del archivo de una ruta y os.path.dirname()devuelve el directorio base.

```py
ruta = "/ruta/carpeta/archivo.txt"
nombre_archivo = os.path.basename(ruta)
directorio_base = os.path.dirname(ruta)

print("Nombre de archivo:", nombre_archivo)
print("Directorio base:", directorio_base)
```
### Dividir Ruta en Directorio y Nombre de Archivo:
`os.path.split()` divide una ruta en el directorio base y el nombre de archivo.

```py
ruta = "/ruta/carpeta/archivo.txt"
directorio, nombre_archivo = os.path.split(ruta)

print("Directorio:", directorio)
print("Nombre de archivo:", nombre_archivo)
```

#### Verificar si existe una ruta:
`os.path.exists()` comprueba si existe una ruta dada en el sistema.

```py
ruta = "/ruta/carpeta/archivo.txt"
if os.path.exists(ruta):
    print("La ruta existe")
else:
    print("La ruta no existe")
```

### Crear directorios anidados:
`os.makedirs()` crea directorios anidados en una ruta dada.

```py
nueva_ruta = "/nuevo/directorio/anidado"
os.makedirs(nueva_ruta)
```

#### Listar Archivos en un Directorio:
`os.listdir()` devuelve una lista de nombres de archivos y carpetas en un directorio.

```py
directorio = "/ruta/del/directorio"
archivos = os.listdir(directorio)
print("Archivos en el directorio:", archivos)
```
### Verificar si es Archivo o Directorio:
`os.path.isfile()` verifica si una ruta dada es un archivo, y `os.path.isdir()` verifica si es un directorio.

```py
ruta = "/ruta/carpeta"
if os.path.isdir(ruta):
    print("Es un directorio")
elif os.path.isfile(ruta):
    print("Es un archivo")
else:
    print("No es un archivo ni un directorio")
```
### Cambiar el Directorio Actual:
`os.chdir()` cambia el directorio actual.

```py
nuevo_directorio = "/nuevo/directorio"
os.chdir(nuevo_directorio)
```

### Eliminar un Archivo o Directorio:
`os.remove()` eliminar un archivo, y os.rmdir()eliminar un directorio vacío.

```py
archivo_a_eliminar = "/ruta/archivo_a_eliminar.txt"
os.remove(archivo_a_eliminar)

directorio_a_eliminar = "/ruta/directorio_a_eliminar"
os.rmdir(directorio_a_eliminar)
```

Estos son solo algunos ejemplos de cómo puedes utilizar el módulo osen Python para manejar rutas del sistema
