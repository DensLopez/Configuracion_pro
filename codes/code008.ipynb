El primer paso es importar la librería con `import pathlib`.

Después tendremos que crear un objeto `Path` para poder manipular las rutas según nuestro S.O. Así que lo crearemos de la siguiente forma: `pathlib.Path`, nos mostrará la siguiente salida:



El `.` significa la ruta en la que estoy, es decir `aquí`, pero si yo quiero saber de manera extendida la ruta, lo indico con `resolve()`.



Después lo guardo en una variable. Y accedo a la ruta padre, después accedo a una carpeta en especifico.

Para realizar operaciones con el objeto:

```py
# Crear un directorio
DATA_DIR.joinpath('pathlib').mkdir()
# Listar elementos de un directorio
list(DATA_DIR.glob('*'))
# Listar elementos con patron especifico
list(DATA_DIR.glob('.git*'))
```
